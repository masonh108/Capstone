```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
```

```python
URL = 'https://en.wikipedia.org/wiki/FLOPS'
```

```python
response = requests.get(URL)
```

```python
response.status_code
```

```python
soup = BS(response.text)
```

```python
print(soup.prettify())
```

```python
cost_of_computing = pd.read_html(str(soup.findAll('table', attrs = {'class' : 'wikitable'})))[2]
```

```python
cost_of_computing.columns = cost_of_computing.columns.droplevel(0)
```

```python
cost_of_computing = cost_of_computing.drop(columns = {'Platform providing the lowest cost per GFLOPS', 'Comments'}).rename(columns = {'Date' : 'year'})
```

```python
cost_of_computing['year'] = cost_of_computing['year'].str[-4:]
```

```python
cost_of_computing
```

```python
cost_of_computing.to_csv('../data/cost_of_computing.csv')
```
